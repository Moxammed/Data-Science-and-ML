```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

x=[1,2,3,4,5]
y=[10,20,25,30,40]

plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
names = ['A', 'B', 'C']
scores = [70, 85, 90]

plt.bar(names, scores)
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Scores")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [5, 7, 4, 8, 6]

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
df = pd.DataFrame({
    'Age': [23, 33, 44],
    'Salary': [1200, 1400, 6000]
})

df.plot(x='Age', y='Salary')
plt.title("Line Plot")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

ar=np.arange(0,11)
plt.plot(ar,ar**2)
plt.show()
```

```python
#crate a bar chart showing counts of 5 random categories
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data={'Category':['A','B','C','D','E'],
      'Values':[10,15,7,12,9]}

df=pd.DataFrame(data)
plt.bar(df['Category'],df['Values'])
plt.show()
```

```python
#Plot a histogram of 1000 random numbers from a normal distribution.
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

ar = np.random.randn(10000000)
plt.hist(ar)
plt.show()
```

```python
#4.	Scatter plot of two random arrays and color points by a third random array.

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

x=np.random.rand(50)
y=np.random.rand(50)
color= np.random.randint(1,101,50)

plt.scatter(x,y,c=color)
plt.show()
```

```python




#5.	Use Seaborn to plot a heatmap of correlation matrix from a random 44 DataFrame.
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np


arr=np.random.randint(1,101,(4,4))
df=pd.DataFrame(arr,columns=['A','B','C','D'])
sns.heatmap(df.corr(),annot=True)
plt.show()
```
