```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

X = np.array([[1,2], [3,4], [5,6], [7,8]])
y = np.array([0,0,1,1])
```

```python
# Create the structure
model=Sequential([
    Dense(8, input_shape=(2,), activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
# How the optimizer and weight loss and metrics should connect and work when the training starts
model.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X,y, epochs=50, batch_size=1)
```

```python
# what is the result loss function to reach the result, and what are the metrics and distance to close up
loss,mae = model.evaluate(X,y)
```

```python
# Make predictions
predictions = model.predict(np.array([[9,10]]))
print(predictions)
```
