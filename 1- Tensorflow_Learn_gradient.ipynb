```python
import tensorflow as tf

t = tf.constant([ [1, 2, 3],
                  [4, 5, 6],
                  [7, 8, 9] ])
```

```python
#1.	Extract the second column.

col2= t[:,1]
print(col2)
```

```python
#2.	Extract the first two rows.

twoRows= t[0:2,:]
print(twoRows)
```

```python
#3.	Reverse the rows.

revRows=t[::-1]
print(revRows)
```

```python
#4.	Extract the submatrix:
# 5, 6
# 8, 9

sub=t[1:, 1:]
print(sub)
```

**Tensor Operations**

```python
x = tf.constant([1, 2, 3])
y = tf.constant([4, 5, 6])
```

```python
#1.	Add x and y.
z=x+y

print(z)
```

```python
#2.	Multiply x and y elementwise.

mul=x*y
print(mul)
```

```python
#3.	Compute dot product of x and y.

dotProduct=tf.tensordot(x,y,axes=1)
print(dotProduct)
```

```python
#4.	Compute x squared elementwise.

arrSquared= x**y
print(arrSquared)
```

```python
#5.	Compute mean and standard deviation of y.
arrMean= tf.reduce_mean(y)

yNew= tf.cast(y,tf.float32)
arrSTD = tf.math.reduce_std(yNew)

print(arrMean)
print(arrSTD)
```



**4️⃣ Broadcasting & Reshaping**


```python
#1.	Create a (2,3) tensor and a (3,) tensor. Add them using broadcasting.

twoByThree= tf.random.uniform(shape=(2,3), minval=1,maxval=20, dtype=tf.int32)
three= tf.random.uniform(shape=(3,),minval=1, maxval=100, dtype=tf.int32)

print(twoByThree)
print(three)

print("Addition is as follows")
add= twoByThree + three
print(add)
```

```python
#2.	Reshape a (6,) tensor into (2,3) shape.

six=tf.random.uniform((6,), minval=1, maxval=100, dtype=tf.int32)
print(six)

reshaped=tf.reshape(six,(2,3))
print(reshaped)
```

```python
#3.	Flatten a (3,3) tensor into a 1D tensor.
three=tf.random.uniform((3,3), minval=1, maxval=100, dtype=tf.int32)
print(three)

flattened=tf.reshape(three,[-1])
print(flattened)
```

```python
#4.	Transpose a (2,3) tensor into (3,2).

arrTwo= tf.random.uniform((2,3), minval=1, maxval=100, dtype=tf.int32)
print(arrTwo)

transposed = tf.transpose(arrTwo)
print(transposed)
```

```python
import tensorflow as tf

x = tf.Variable(5.0)

with tf.GradientTape() as tape:
    y = x**2 + 3*x + 2

dy_dx = tape.gradient(y, x)
print(dy_dx)
```
